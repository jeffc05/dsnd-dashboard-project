```python
from sqlite3 import connect
from pathlib import Path
import pandas as pd
```

```python
relative_path = Path('python-package/employee_events/employee_events.db')
db_path = relative_path.absolute()
```

```python
class QueryMixin:
    
    # Define a method named `pandas_query`
    # that receives an sql query as a string
    # and returns the query's result
    # as a pandas dataframe
    def pandas_query(self, sql_query):
        conn = connect(db_path)
        pandas_df = pd.read_sql_query(sql_query, conn)
        conn.close()
        return pandas_df

    # Define a method named `query`
    # that receives an sql_query as a string
    # and returns the query's result as
    # a list of tuples. (You will need
    # to use an sqlite3 cursor)
    def query(self, sql_query):
        conn = connect(db_path)
        cursor = conn.cursor()
        cursor.execute(sql_query)
        results = cursor.fetchall()
        conn.close()
        return results
```

```python
query_mixin = QueryMixin()
```

```python
query_mixin.pandas_query("SELECT name FROM sqlite_master WHERE type='table';")
```

```python
query_mixin.query("SELECT name FROM sqlite_master WHERE type='table';")
```
